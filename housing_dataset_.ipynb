```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Dataset
data = pd.read_excel('/content/housing.csv.xlsx')

# Step 3: Inspect the First Few Rows
print("ðŸ”¹ First 5 rows of the dataset:")
print(data.head())

# Step 4: Basic Information About the Dataset
print("\nðŸ”¹ Dataset Information:")
print(data.info())

# Step 5: Check for Missing Values
print("\nðŸ”¹ Missing Values in Each Column:")
print(data.isnull().sum())

# Step 6: Descriptive Statistics
print("\nðŸ”¹ Descriptive Statistics:")
print(data.describe(include='all'))

# Step 7: Data Types Summary
print("\nðŸ”¹ Data Types:")
print(data.dtypes)

# Step 8: Correlation Matrix
print("\nðŸ”¹ Correlation Matrix (Numerical Columns):")
corr = data.corr(numeric_only=True)
print(corr)

# Step 9: Data Visualization
plt.style.use('seaborn-v0_8')

# 9.1 Histogram for numeric columns
data.hist(figsize=(12, 10), bins=30, edgecolor='black')
plt.suptitle("Distribution of Numeric Features", fontsize=16)
plt.show()

# 9.2 Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap", fontsize=16)
plt.show()

# 9.3 Pairplot for relationships among numeric features
sns.pairplot(data.sample(min(500, len(data))), diag_kind='kde')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()

# 9.4 Boxplot example - If categorical features exist
for col in data.select_dtypes(include='object').columns:
    for num_col in data.select_dtypes(include=np.number).columns:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=col, y=num_col, data=data)
        plt.title(f"{num_col} vs {col}")
        plt.xticks(rotation=45)
        plt.show()

# 9.5 Scatter plot for geographical data (longitude vs latitude)
if {'longitude', 'latitude'}.issubset(data.columns):
    plt.figure(figsize=(10, 6))
    plt.scatter(data['longitude'], data['latitude'],
                c=data['median_house_value'], cmap='viridis', alpha=0.6)
    plt.colorbar(label='Median House Value')
    plt.title("Geographical Distribution of Median House Value")
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.show()

print("\nâœ… EDA Completed Successfully!")
```
